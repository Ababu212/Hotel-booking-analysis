# Data Cleaning for better understanding of data frame

```python
#Adding children and babis in one column
hotel['kid']= (hotel['children']+ hotel['babies']).astype(int)
```

```python
hotel.info()
```

```python
#dropping babies, children, company and agent column
hotel.drop(['babies','children','company','agent'],axis=1,inplace= True) 
```

```python
hotel[(hotel['adults']+hotel['kid'])==0].shape
```

```python
#dropping the rows in which there are 0 values in country
country_nan= hotel[hotel['country']==0]
hotel.drop(country_nan.index, inplace=True)
```

```python
hotel.info()
```

# Dropping of rows in which no guests have arrived

```python
guests= hotel[(hotel['adults']+hotel['kid'])==0]
hotel= hotel.drop(guests.index)
hotel.shape
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
sns.countplot (x= 'arrival_date_year', data= hotel, hue= 'hotel').set_title ('yearly bookings')
#hotelwise yearly booking
```

```python
#Countrywise booking
hotel.groupby('country').adults.sum().sort_values(ascending= False)
```

```python
plt.title("Country wise booking")
ax = sns.countplot(x = "country", data = hotel)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Segmentwise booking
hotel.groupby('market_segment').adults.sum().sort_values(ascending= False)
```

```python
plt.title("Segments wise booking")
ax = sns.countplot(x = "market_segment", data = hotel)
plt.xticks(rotation = 90)
plt.show()
#market segmentwise booking
```

```python
plt.title("Most preferred mealtype by customers")
ax = sns.countplot(x = "meal", data = hotel)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.title("Busiest month of hotel booking")
ax = sns.countplot(x = "arrival_date_month", data = hotel)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.title("Preferred room type by customers")
ax = sns.countplot(x = "reserved_room_type", data = hotel)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.title ('Reserved room type vs assigned room type')
sns.barplot(x= 'reserved_room_type' ,y = "assigned_room_type", data = hotel)
```
